## 5. Slicing
Slicing is used to access a "part" or "chunk" of a string.

**Syntax:** `str[starting_index : ending_index]`
* The `ending_index` is **not included** (it stops one index before).

````python
str = "Apna College"

print(str[0:4])   # Output: Apna (0 to 3)
print(str[5:12])  # Output: College (5 to 11)
print(str[5:len(str)]) # Output: College

# Shorthand Slicing
print(str[:4])    # Same as str[0:4]
print(str[5:])    # Same as str[5:end]

# Negative Indexing (Backward Counting)
# Apple: A(0), p(1), p(2), l(3), e(4)
# Apple: A(-5), p(-4), p(-3), l(-2), e(-1)
str_apple = "Apple"
print(str_apple[-3:-1]) # Output: pl
````

# Python Full Course: Lecture 3 - Deep Dive into Lists & Tuples

## 1. Understanding Lists
A **List** is a built-in data type that stores a collection of items in a single variable. Items in a list are ordered, changeable (mutable), and allow duplicate values.

### Why use Lists?
Imagine you have 100 students. Creating variables like `marks1`, `marks2` ... `marks100` is impossible to manage. A list allows you to store all 100 values in one place: `marks = [85, 92, 78, ...]`.

### Key Characteristics:
1. **Ordered:** The items have a defined order that will not change unless you specifically change it.
2. **Mutable:** You can add, remove, or change items after the list is created.
3. **Dynamic:** A single list can hold different types of data (Mixed Data Types).

````python
# Creating different types of lists
numbers = [10, 20, 30, 40]
names = ["Shradha", "Aman", "Karan"]
mixed_list = ["Python", 2024, 98.5, True, None]

print("Number List:", numbers)
print("Mixed List:", mixed_list)

# Finding the size of a list
print("Total items in Mixed List:", len(mixed_list))
````

## 2. List Indexing (Accessing Data)
Every item in a list has a specific position number called an **index**.

### Positive Indexing (Left to Right):
Starts from **0** up to **len-1**.
| Index | 0 | 1 | 2 | 3 |
|---|---|---|---|---|
| Value | "Apple" | "Banana" | "Cherry" | "Date" |

### Negative Indexing (Right to Left):
Starts from **-1** (the last item).
| Index | -4 | -3 | -2 | -1 |
|---|---|---|---|---|
| Value | "Apple" | "Banana" | "Cherry" | "Date" |

````python
fruits = ["Apple", "Banana", "Cherry", "Date"]

print("First item (Index 0):", fruits[0])
print("Last item (Index -1):", fruits[-1])
print("Second to last (Index -2):", fruits[-2])
````

## 3. Mutability (The Core Power of Lists)
Unlike Strings, Lists are **Mutable**. This means we can reach into the memory and swap a value without creating a whole new list.

**Memory Concept:**
When you change `marks[0] = 100`, Python goes to the exact address of that first item and replaces it.

````python
student_info = ["Karan", 85, "Mumbai"]
print("Before Change:", student_info)

# Swapping data
student_info[1] = 98 # Updated marks
student_info[2] = "Delhi" # Updated city

print("After Change:", student_info)
````

## 4. Comprehensive List Slicing
Slicing allows you to create a "sub-list" from an existing list.
**Formula:** `list_name[start : stop : step]`

* `start`: Index where the slice begins (Inclusive).
* `stop`: Index where the slice ends (Exclusive - stops one before).
* `step`: Optional jump value (default is 1).

````python
nums = [10, 20, 30, 40, 50, 60, 70, 80]

print("Basic slice (1 to 4):", nums[1:5]) # [20, 30, 40, 50]
print("Starting from 0 to 3:", nums[:4]) # [10, 20, 30, 40]
print("From 4 to the end:", nums[4:])    # [50, 60, 70, 80]

# Using the Step parameter
print("Every 2nd number:", nums[::2])     # [10, 30, 50, 70]

# Reversing a list using slicing
print("Reversed List:", nums[::-1])
````

## 5. List Methods (The Cheat Sheet)
These are essential tools for managing list data efficiently.

| Method | Action | Example |
|---|---|---|
| `append(val)` | Adds item at the **end** | `L.append(4)` |
| `insert(idx, val)` | Adds item at a **specific index** | `L.insert(1, "hi")` |
| `sort()` | Sorts list in **ascending order** | `L.sort()` |
| `sort(reverse=True)` | Sorts list in **descending order** | `L.sort(reverse=True)` |
| `reverse()` | Reverses the current order | `L.reverse()` |
| `remove(val)` | Removes the **first occurrence** of val | `L.remove(2)` |
| `pop(idx)` | Removes and returns item at **index** | `L.pop(0)` |
| `clear()` | Removes **all** items from list | `L.clear()` |

````python
my_list = [4, 2, 8, 1, 2, 9]

my_list.append(10)
print("Append 10:", my_list)

my_list.sort()
print("Sorted:", my_list)

my_list.remove(2) # Only the FIRST 2 is removed
print("Remove 2:", my_list)

deleted_val = my_list.pop(0) # Removes item at index 0
print("Popped item:", deleted_val)
print("List after pop:", my_list)
````

## 6. Understanding Tuples
A **Tuple** is a sequence of immutable Python objects. Tuples are sequences, just like lists. The main difference is that tuples cannot be changed (Immutable) and use parentheses `()`.

### Why use Tuples if Lists are better?
1. **Performance:** Tuples are faster than lists for small data sets.
2. **Safety:** If you have data that should never change (like API keys or Pi value), a tuple prevents accidental deletion or modification.
3. **Dictionary Keys:** Only immutable types (like tuples) can be used as keys in a dictionary (Chapter 4).

````python
tup = (10, 20, 30, 40)
print("Tuple Value:", tup[0])

# Attempting to change a value will fail:
try:
    tup[0] = 50
except TypeError as e:
    print("Error:", e)

## IMPORTANT: Single Element Tuple
tup1 = (5)   # This is seen as an INTEGER by Python
tup2 = (5,)  # This is a TUPLE (Note the comma)
print("tup1 type:", type(tup1))
print("tup2 type:", type(tup2))
````

## 7. Tuple Methods
Since tuples are immutable, they only have two primary methods:

1. **`count(val)`**: Returns the number of times a value occurs.
2. **`index(val)`**: Returns the first index where the value is found.

````python
data = (1, 2, 3, 2, 4, 2, 5)
print("Occurrences of 2:", data.count(2)) # Output: 3
print("First index of 4:", data.index(4)) # Output: 4
````

## 8. Summary Comparison

| Feature | List | Tuple |
|---|---|---|
| **Brackets** | `[ ]` | `( )` |
| **Mutability** | Mutable (Changeable) | Immutable (Constant) |
| **Speed** | Slower | Faster |
| **Methods** | Many (append, pop, etc.) | Few (count, index) |
| **Usage** | Frequent changes | Read-only data |

# Python Full Course: Lecture 4 - Dictionaries & Sets

## 1. What is a Dictionary?
A **Dictionary** is a built-in data type that stores data in **Key: Value** pairs.

* **Real-world Analogy:** Just like a language dictionary where you look up a **Word** (Key) to find its **Meaning** (Value).
* **Syntax:** Use curly braces `{ }` with keys and values separated by a colon `:`.

### Key Properties:
1. **Unordered:** There is no fixed index (like 0, 1, 2) for items.
2. **Mutable:** You can change values and add/remove keys after creation.
3. **No Duplicates:** You cannot have two keys with the same name. If you try, the old value will be overwritten.

````python
info = {
    "key" : "value",
    "name" : "apna college",
    "learning" : "coding",
    "age" : 35,
    "is_adult" : True,
    "marks" : 94.4
}

print(info)
print(type(info))
````

## 2. Accessing & Modifying Dictionaries
To get a value, you call its **Key**. To change it, you simply re-assign a value to that key.

````python
student = {
    "name" : "Rahul Kumar",
    "marks" : 85,
    "subject" : "Python"
}

# Accessing
print(student["name"]) # Rahul Kumar

# Modifying
student["name"] = "Neha Kumar"
student["marks"] = 99

# Adding a new Key-Value pair
student["surname"] = "Singh"

print(student)
````

## 3. Nested Dictionaries
A dictionary can contain another dictionary as a value. This is used for complex data like student records with individual subject marks.

````python
student = {
    "name" : "Rahul Kumar",
    "subjects" : {
        "phy" : 97,
        "chem" : 95,
        "math" : 98
    }
}

print(student["subjects"])         # Prints inner dictionary
print(student["subjects"]["phy"])  # Prints specific nested value: 97
````

## 4. Dictionary Methods
| Method | Action |
|---|---|
| `keys()` | Returns all keys |
| `values()` | Returns all values |
| `items()` | Returns all (key, value) pairs as tuples |
| `get("key")` | Returns the value (Safe method: returns None if key not found, instead of Error) |
| `update(new_dict)` | Inserts specified items to dictionary |

````python
student = {"name": "Rahul", "age": 20}

print(student.keys())      # dict_keys(['name', 'age'])
print(student.values())    # dict_values(['Rahul', 20])
print(list(student.items())) # [('name', 'Rahul'), ('age', 20)]

print(student.get("name")) # Rahul
print(student.get("name2"))# None (Safe)

student.update({"city" : "Delhi", "age" : 21})
print(student)
````

## 5. What are Sets?
A **Set** is an unordered collection of unique items.

* **Unique:** No duplicate items allowed. If you add the same item twice, it only stores it once.
* **Immutable Elements:** You can store strings, numbers, and tuples in a set, but NOT lists or dictionaries (because they can change).
* **Set itself is Mutable:** You can add/remove items from the set.

````python
nums = {1, 2, 3, 4, 4, "hello", "hello"}
print(nums)  # Output will only show {1, 2, 3, 4, 'hello'}
print(len(nums)) # 5 items

## Creating an EMPTY Set
collection = set() # Correct way
wrong_way = {}     # WRONG: This creates an empty Dictionary
````

## 6. Set Methods
* `add(el)` : Adds an element.
* `remove(el)` : Removes an element.
* `pop()` : Removes a random value.
* `clear()` : Empties the set.
* `union(set2)` : Combines two sets (returns unique items from both).
* `intersection(set2)` : Returns only common items from both sets.

````python
set1 = {1, 2, 3}
set2 = {2, 3, 4}

print(set1.union(set2))        # {1, 2, 3, 4}
print(set1.intersection(set2)) # {2, 3}
````

# Python Full Course: Lecture 5 - Master Class on Loops

## 1. What are Loops?
Loops are used to repeat a block of code multiple times as long as a specific condition is met.

### Why do we need them?
* **Automation:** If you want to print "Hello" 100,000 times, writing `print` manually is impossible.
* **Data Processing:** Iterating through lists of millions of users to send emails.
* **Efficiency:** Keeps code DRY (Don't Repeat Yourself).

---

## 2. The `while` Loop
The `while` loop is **condition-based**. It continues to execute as long as the condition remains **True**.

### Syntax:
```python
while condition:
    # code to execute
```

### The Three Pillars of a While Loop:
1. **Initialization:** Setting the starting point (e.g., `i = 1`).
2. **Condition:** The rule that stops the loop (e.g., `i <= 5`).
3. **Update (Increment/Decrement):** Changing the variable so the loop eventually ends (e.g., `i += 1`).

````python
count = 1
while count <= 5:
    print("Hello World", count)
    count += 1

print("Loop has ended.")
````

### ⚠️ The Danger: Infinite Loops
If the condition never becomes `False`, the loop will run forever until your computer runs out of memory or the program crashes.

**Example of what NOT to do:**
```python
i = 1
while i > 0:
    print(i)
    # Missing i += 1 or i -= 1 makes this run forever!
```

---

## 3. The `for` Loop
The `for` loop is used for **Sequential Traversal**. It is best when you want to go through every item in a List, Tuple, String, or Dictionary.

### Syntax:
```python
for element in collection:
    # code to execute
```

````python
nums = [1, 2, 3, 4, 5]

print("Traversing a List:")
for val in nums:
    print(val)

print("\nTraversing a String:")
name = "Python"
for char in name:
    print(char)
````

### Optional `else` with Loops
Python allows an `else` block after a loop. It executes **only after the loop completes normally**. If the loop is terminated by a `break` statement, the `else` block is skipped.

````python
str = "apna college"
for char in str:
    print(char)
else:
    print("Loop completed successfully.")
````

---

## 4. Range Function: `range()`
The `range()` function returns a sequence of numbers, starting from 0 by default, and increments by 1 (by default), and stops before a specified number.

### Parameters:
`range(start, stop, step)`
* **start:** Optional (default 0). Included.
* **stop:** Mandatory. **Excluded** (stops at stop-1).
* **step:** Optional (default 1). The gap between numbers.

````python
print("Stop only - range(5):", list(range(5)))
print("Start & Stop - range(1, 5):", list(range(1, 5)))
print("With Step - range(1, 10, 2):", list(range(1, 10, 2)))

# Practical use in for loop
for i in range(100, 105):
    print("Processing index:", i)
````

---

## 5. Loop Control Statements

### A. `break`
Used to terminate the loop immediately. Use this when you found what you were looking for and don't need to check the rest of the items.

### B. `continue`
Skips the rest of the current iteration and jumps to the next one. Use this to "skip" specific values (like skipping even numbers).

### C. `pass`
A null statement. It does nothing. Used as a placeholder when you haven't written the logic yet but want to avoid syntax errors.

````python
print("--- Testing Break (stop at 3) ---")
for i in range(1, 6):
    if i == 3:
        break
    print(i)

print("\n--- Testing Continue (skip 3) ---")
for i in range(1, 6):
    if i == 3:
        continue
    print(i)

print("\n--- Testing Pass ---")
for i in range(5):
    pass # placeholder
print("I will write code inside that loop later!")
````

## **1. What are Functions?**

A **Function** is a block of statements (a group of code) that performs a specific task.

### **Why do we use Functions?**

Imagine you need to calculate the sum of two numbers multiple times in your code. Without functions, you would have to write the addition logic again and again. This leads to **Redundancy** (repeating the same thing).
Functions help us follow the **DRY (Don't Repeat Yourself)** principle.

### **Syntax of a Function**

To create a function, we use the `def` keyword (short for define).

````python
# Defining the function
def function_name(parameter1, parameter2):
    # logic of the function
    return result

# Calling the function
function_name(arg1, arg2)
````

---

## **2. Basic Function Examples**

Let's look at a simple function that calculates the sum of two numbers.

### **Code Example: Simple Addition Function**

````python
# Function definition
def calc_sum(a, b):
    sum_val = a + b
    print(f"The sum is: {sum_val}")
    return sum_val

# Function call 1
calc_sum(5, 10)

# Function call 2
calc_sum(12, 17)

# Function call 3
calc_sum(20, 25)
````

### **Explanation of Terms:**

1. **Parameters:** The variables used in the function definition (like `a` and `b`).
2. **Arguments:** The actual values passed to the function when calling it (like `5` and `10`).
3. **Return:** The value that the function "gives back" to the place where it was called.

## **4. The `return` Statement**

The `return` keyword is used to send a result back. If a function does not have a return statement, it returns `None` by default.

````python
def multiply(a, b):
    return a * b

result = multiply(4, 5)
print("The multiplication result is:", result)
````

## **5. Default Parameters**

Sometimes, we want a parameter to have a "backup" value if the user doesn't provide one. These are called **Default Parameters**.

**Note:** Default parameters must always be at the end of the parameter list.

````python
def greet(name, message="Good Morning"):
    print(f"Hello {name}, {message}")

greet("Shradha") # Uses default message
greet("Shradha", "Good Night") # Overrides default message
````

## **7. Recursion**

**Recursion** is when a function calls itself. It is a way to solve a problem by breaking it down into smaller, similar sub-problems.

### **Important Components of Recursion:**

1. **Recursive Step:** Calling the function again with a smaller value.
2. **Base Case:** The condition where the function stops calling itself. **Without a base case, the function will run forever (Infinite Recursion).**

### **Example 1: Factorial using Recursion**

The factorial of `n` (written as `n!`) is `n * (n-1) * (n-2)... * 1`.

````python
def factorial(n):
    # Base Case
    if n == 0 or n == 1:
        return 1
    # Recursive Step
    return n * factorial(n - 1)

print("Factorial of 5 is:", factorial(5))
````

## **8. Practice Questions (From the Video)**

### **Q1: Write a function to print the length of a list.**

````python
cities = ["Delhi", "Mumbai", "Chennai", "Pune", "Kolkata"]

def print_len(list_input):
    print("Length of list is:", len(list_input))

print_len(cities)
````

### **Q2: Write a function to print elements of a list in a single line.**

````python
def print_list(list_input):
    for item in list_input:
        print(item, end=" ")
    print() # To add a newline at the end

print_list(cities)
````

### **Q3: Write a function to find the factorial of n using a Loop (Iterative).**

````python
def calc_fact(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    print("Iterative Factorial:", fact)

calc_fact(5)
````

### **Q4: Write a recursive function to calculate the sum of first n natural numbers.**

````python
def sum_natural(n):
    if n == 0:
        return 0
    return n + sum_natural(n - 1)

print("Sum of first 5 natural numbers:", sum_natural(5))
````

## **9. Advanced: Recursion with Lists**

Printing every element of a list using recursion.

````python
def print_list_recursive(list_input, index=0):
    # Base Case: stop when index reaches length of list
    if index == len(list_input):
        return

    # Logic
    print(list_input[index])

    # Recursive Step
    print_list_recursive(list_input, index + 1)

fruits = ["Mango", "Litchi", "Apple", "Banana"]
print_list_recursive(fruits)
````

## **10. Additional Concepts (Bonus Content)**

### **Variable Scope**

Variables defined inside a function are **Local Variables** (only accessible inside that function). Variables defined outside are **Global Variables**.

````python
x = 10 # Global

def my_func():
    y = 5 # Local
    print("Inside function, x =", x)
    print("Inside function, y =", y)

my_func()
# print(y) # This would give an error because y is local to the function.
````

### **Lambda Functions (One-liner functions)**

For very small tasks, Python offers "Lambda" functions.

````python
# Normal way
def square(n):
    return n*n

# Lambda way
square_lambda = lambda n: n*n

print("Square of 4 is:", square_lambda(4))
````
